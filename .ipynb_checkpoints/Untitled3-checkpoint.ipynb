```python
raw_ds=pd.read_csv('tmdb_5000_movies.csv')
```

```python
raw_ds.head()
raw_ds.info()
name_col=['budget','genres','original_language','production_countries','title','popularity','revenue','status',
          'release_date','vote_average','vote_count']
film=raw_ds[name_col]
film.head(3)
```

```python
# словарь жанров
genres_dict={}
genres_dict.clear()
for key, value in raw_ds['genres'].items():
    js.loads(value)
   # print(type(js_keyword))
    for i in js.loads(value): 
        genres_dict[i['id']]=i['name']  #print (i['id'], i['name'])
print(genres_dict)
```

```python
# колонки для жанров
for key,value in genres_dict.items():
    film[value]=0
film.info()
```

```python
#заполняем столбцы жанров, скорее всего это можно сделать красивее ((
for key, value in film['genres'].items():
    for i in js.loads(value):
        #print(key,i['id'], i['name'])
        film[i['name']].loc[key] = 1       
            
```

```python
contry_dict={}
contry_dict.clear()
country_rate = {}
country_rate.clear()
for key, value in raw_ds['production_countries'].items():
    js.loads(value)
   # print(type(js_keyword))
    for i in js.loads(value): 
        contry_dict[i['iso_3166_1']]=i['name']  #print (i['id'], i['name'])
        country_rate[i['iso_3166_1']]=country_rate.get(i['iso_3166_1'], 0)+1
#print(contry_dict)
print(country_rate)
```

```python
#колонки для стран-производителей фильмов
for key, value in country_rate.items():
    if value >= 50 :
        film[key]=0
film.info()
```

```python
# заполняем колонки стран-производителей
for key, value in film['production_countries'].items():
    for i in js.loads(value):
        if country_rate.get(i['iso_3166_1'])>=50:
            #print(key,i['iso_3166_1'], i['name'])
            film[i['iso_3166_1']].loc[key] = 1 
```

```python
# film страны, участницы

for key, value in film['production_countries'].items():
    film_country = []
    for i in js.loads(value):
        film_country.append(i['iso_3166_1'])
    sort(film_country)
    #print(sort(film_country))
    #film['cntry_names'].loc[key].append(film_country)

def country(x):
    film_country=[]
    for i in js.loads(x):
        film_country.append(i['iso_3166_1'])
    return sort(film_country)

def genres(x):
    film_genres=[]
    for i in js.loads(x):
        film_genres.append(i['name'])
    return sort(film_genres)
```

```python
film['country_name']=film['production_countries'].apply(lambda x: country(x))
```

```python
film['film_genres']=film['genres'].apply(lambda x: genres(x))
```

```python
film['country_n']=film['production_countries'].apply(lambda x: len(js.loads(x)))
```

```python
film['genres_n']=film['genres'].apply(lambda x: len(js.loads(x)))
```
